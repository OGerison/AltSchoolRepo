We have implemented pandas functions to our dataset gotten from Kaggle https://www.kaggle.com/datasets/rishikeshkonapure/hr-analytics-prediction/ to help us show you the power of pandas.

```python
# Importing the pandas library we need
import pandas as pd
```

```python
# Reading the dataset into a dataframe
file_path = "E:\AltSchoolofData\Semester 1\Projects\HR-Employee-Attrition.csv"
df = pd.read_csv(file_path)
```

```python
# Finding the number of rows and columns in our dataframe
df.shape
```

```python
# We can use head to see the first 5 rows of out dataframe
df.head()
```

```python
# Or tail to see the last 5 rows.
df.tail()
```

```python
# Finding out the columns we have in our dataframe
df.columns
```

```python
# Describing the data to understand and see how it is distributed.
df.describe()
```

```python
# Checking the data types.
df.dtypes
```

```python
# Checking for the number of null values in each column
df.isna().sum()
```

Our dataframe did not have null values so we did not use df.dropna() to drop such values or df.fillna() to fill the missing values.

```python
# We decided to drop certain columns from our dataframe
df.drop(["StockOptionLevel", "JobInvolvement"], inplace=True, axis=1)
```

```python
#Rename the age column since its current name did not make sense.
df.rename(columns={"Ã¯Â»Â¿Age": "Age"}, inplace=True)
```

```python
# Using Valuecounts to find out how much the employees travelled.
df["BusinessTravel"].value_counts()
```

```python
# But when we add normalize, it gives the value_counts a max value of 1 so they are like in the same range
df["BusinessTravel"].value_counts(normalize=True)
```

```python
# Finding out the mean of a column
df["Age"].mean()
```

```python
# Finding the unique values in a column like a set picks unique values only
df.Age.unique()
```

```python
# Using groupby to find average daily rate by department
df.groupby(by="Department")["DailyRate"].mean()
```

```python
# Using a pivot table to find out the average distance from home as per their educational fields.
df.pivot_table(values="DistanceFromHome", index = "EducationField", aggfunc="mean")
```

```python
# Instead of using df.head to print the first 5, lets use iloc to print the first 20 rows
df.iloc[:21]
```

```python
# But we can also use loc to print the first 20 rows for a column(s) we want to see based on a condition.
df.loc[df["Age"] < 30][["Age","DailyRate"]]
```

```python
# We can use groupby in conjuction with get group to separate people in the sales department in a df of their own.
df.groupby(by="Department").get_group("Sales")
```

```python
# We can also do some visualizing in this case the distribution of the number of people in the diffrent departments
df["Department"].value_counts().plot(kind="barh", title="Distribution by Department", ylabel="Department",xlabel="Count")
```
